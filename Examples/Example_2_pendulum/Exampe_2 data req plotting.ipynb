```python
xls = pd.ExcelFile('pendulum_data_requirement.xlsx')

df_dict = {sheet_name: xls.parse(sheet_name) for sheet_name in xls.sheet_names}
```

```python
plt.figure()
marker_list = ['o', '^', 'o']
color_list = ['#332288', '#332288', '#88CCEB']
for key, marker, color in zip(df_dict, marker_list, color_list):
    if key == "Pendulum experiment video":
        continue
    df = df_dict[key]
    plt.plot(df["Polynomial Degree"][:4], df["num data points"][:4], marker=marker, color=color, label=key)

plt.title('Polynomial Degree vs Num Data Points')
plt.xlabel('Polynomial Degree')
plt.ylabel('Num of Frames Per Footage')
plt.xticks([2, 3, 4, 5])
plt.grid(True)
plt.legend()
plt.savefig('Example_2_data_requirement.svg', format='svg')
plt.show()

```
